```python
 #* PTS EQUIPOS JORNADAS:
def pts_equipos():
    palop = int(input('Pts Palop: '))
    fale = int(input('Pts Fale: ')) 
    tony = int(input('Pts Tony: '))
    lope = int(input('Pts Lope: '))
    ruso = int(input('Pts Ruso: '))
    coquina = int(input('Pts Coquina: '))
    papu = int(input('Pts Papu: ')) 
    kike = int(input('Pts Kike: '))
    gonzo = int(input('Pts Gonzo: '))
    diego = int(input('Pts Diego: '))
    kero = int(input('Pts Kero: ')) 
    armada = int(input('Pts Armada: '))

    return palop, fale, tony, lope, ruso, coquina, papu, kike, gonzo, diego, kero, armada

```

```python
def actualizar_puntos(df_jornada, pts_jornada):

    #* Actualizar el DataFrame con los puntos de cada equipo
    #! MODIFICAR CON LAS POSICIONES CORRECTAS DE LOS EQUIPOS EN LA JORNADA:
    jornada_1.loc[jornada_1['LOCAL'] == 'Palop', 'RTDO L'] = pts_jornada[0]
    jornada_1.loc[jornada_1['VISITANTE'] == 'Fale', 'RTDO V'] = pts_jornada[1]
    jornada_1.loc[jornada_1['LOCAL'] == 'Tony', 'RTDO L'] = pts_jornada[2]
    jornada_1.loc[jornada_1['VISITANTE'] == 'Lope', 'RTDO V'] = pts_jornada[3]
    jornada_1.loc[jornada_1['VISITANTE'] == 'Ruso', 'RTDO V'] = pts_jornada[4]
    jornada_1.loc[jornada_1['VISITANTE.1'] == 'Coquina', 'RTDO V.1'] = pts_jornada[5]
    jornada_1.loc[jornada_1['LOCAL.1'] == 'Papu', 'RTDO L.1'] = pts_jornada[6]
    jornada_1.loc[jornada_1['LOCAL.1'] == 'Kike', 'RTDO L.1'] = pts_jornada[7]
    jornada_1.loc[jornada_1['VISITANTE.1'] == 'Gonzo', 'RTDO V.1'] = pts_jornada[8]
    jornada_1.loc[jornada_1['LOCAL.1'] == 'Diego', 'RTDO L.1'] = pts_jornada[9]
    jornada_1.loc[jornada_1['LOCAL'] == 'Kero', 'RTDO L'] = pts_jornada[10]
    jornada_1.loc[jornada_1['VISITANTE.1'] == 'Armada', 'RTDO V.1'] = pts_jornada[11]

    return df_jornada
```

```python
def actualizar_clasificacion(df, columna_1x2, clasificacion, col_local = 'LOCAL', col_visitante = 'VISITANTE'):
  """
  Actualiza una clasificación de equipos basada en los resultados de partidos.

  Args:
    df: DataFrame con los resultados de los partidos.
    columna_1x2: Nombre de la columna en df que contiene los resultados (1, X, 2).
    clasificación: Clasificación anterior a la jornada.

  Returns:
    DataFrame con la clasificación actualizada.

  Nota: 
    Cada vez que se activa la función realiza la suma según los datos de los argumentos de la función
  """

  #* Iterar sobre los partidos por cada fila y por cada columna:
  for _, fila in df.iterrows(): #* pone una _ para que esa variable no se tenga en cuenta porque no se va a usar
    ganador = fila[columna_1x2] #* Obtine el 1X2 de cada fila
    equipo_local = fila[col_local]  #* Obtiene el equipo local de cada fila
    equipo_visitante = fila[col_visitante]  #* Obtiene el equipo visitante de cada fila

    #* Actualizar la clasificación según el resultado:
    if ganador == 1:
      clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'Puntos'] += 3
      clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'G'] += 1
      clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'P'] += 1
    elif ganador == 'X':
      clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'Puntos'] += 1
      clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'E'] += 1
      clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'Puntos'] += 1
      clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'E'] += 1
    else:
      clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'P'] += 1
      clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'Puntos'] += 3
      clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'G'] += 1

    # Actualizar el número de partidos jugados
    clasificacion.loc[clasificacion['Equipos'] == equipo_local, 'PJ'] += 1
    clasificacion.loc[clasificacion['Equipos'] == equipo_visitante, 'PJ'] += 1

  # Ordenar la clasificación por puntos de forma descendente
  clasificacion = clasificacion.sort_values(by=['Puntos', 'Nºs 1º', 'Max.pts'], ascending=[False, False, False])
  clasificacion = clasificacion.reset_index(drop=True)
  
  return clasificacion
```
